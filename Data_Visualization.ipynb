```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,6)
y=x**2
```

**Line Plot-Stock Price Trends Over Time**

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.show()
```

**Scatter plot -Correlation b/w study hours and marks**

```python
plt.scatter(x,y,color="blue")
plt.title("Scatter plot")
plt.show()
```

**Bar Plot -Compare the number values like sales,profit**

```python
categories=['A','B','C']
values=[5,7,8]
plt.bar(categories,values)
plt.title("Bar plot")
plt.show()
```

**Horizontal bar chart-used to compare the ranking,score,runs**

```python
plt.barh(categories,values,color="green")
plt.title("Horizintal Bar Chart")
plt.show()
```

**Histogram-used to distribution of exam scores**

```python
data=np.random.randn(1000)
plt.hist(data,bins=20,color="gray",edgecolor="black")
plt.title("Histogram")
plt.show()
```

**Pie chart market share of companies**


```python
sizes=[30,45,50,60,70]
lab=['A','B','C','D','E']
plt.pie(sizes,labels=lab,autopct='%1.1f%%') #autopct=adjust size handle with float points
plt.title("Pie Chart")
plt.show()
```

**Stack Plot- used to compare usage components like electric usage**

```python
days=[1,2,3,4,5,6]# x values
A=[1,2,3,4,5,6]
B=[1,1,2,2,3,3]
plt.stackplot(days,A,B,labels=['A','B'])
plt.legend()
plt.title("Stack plot")
plt.show()
```

```python
plt.fill_between(x,y,color="skyblue",alpha=0.6) #Alpha" is a measure of opacity, with values and filling area
plt.plot(x,y,color='black')
plt.title("Fill Between")
plt.show()
```

**Multiple lines _to compare two numerical values**

```python
plt.plot(x,y,label="x^2")
plt.plot(x,x^3,label="3x")
plt.legend()
plt.title("Multiple lines")
plt.show()
```

**Subplot-Create dashboard with multiple plots**

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

axs[0].plot(x, y)
axs[0].set_title("y = x^2")

axs[1].bar(x, x**3)
axs[1].set_title("y = x^3")

plt.tight_layout()
plt.show()
```

**Seaborn-Visualization**

```python
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

**Scatterplot-Relationship b/w income and spending**

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

**Line plot-wensite traffic over a week**

```python
sns.lineplot(x='size',y='tip',data=df)
```

**Bar plot-Avg salary ny department**

```python
sns.barplot(x='day',y='total_bill',data=df)
```

**Count plot- Number of students per grade**

```python
sns.countplot(x='day',data=df,color='yellow')
```

**Boxplot-compare the day and total_bill**

```python
sns.boxplot(x='day',y='total_bill',data=df,color='green')
```

**Violin Plot -Distribution of House prices by city**

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='day',y='tip',data=df,color='orange')
```

**Strip plot-Individual Customer Spending by day**

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

Swarm polt- Individual marks with no overlap

```python
sns.swarmplot(x='day',y='total_bill',data=df,color='purple')
```

**HeatMap-Correlation b/w Variables in sales**

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
```

**Pair Plot-Explore all variable relationship in dataset**

```python
sns.pairplot(df)
```
